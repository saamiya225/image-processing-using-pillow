```python
import cv2
from PIL import Image
```

```python
#import all threats images
threat1 = Image.open("../BaggageAI_CV_Hiring_Assignment/threat_images/BAGGAGE_20170522_113049_80428_A.jpg")
threat2 = Image.open("../BaggageAI_CV_Hiring_Assignment/threat_images/BAGGAGE_20170522_115645_80428_B.jpg")
threat3 = Image.open("../BaggageAI_CV_Hiring_Assignment/threat_images/BAGGAGE_20170523_085803_80428_D.jpg")
threat4 = Image.open("../BaggageAI_CV_Hiring_Assignment/threat_images/BAGGAGE_20170523_094231_80428_B.jpg")
threat5 = Image.open("../BaggageAI_CV_Hiring_Assignment/threat_images/BAGGAGE_20170524_075554_80428_B.jpg")
```

```python
#import all the background images
bg1 =  Image.open("../BaggageAI_CV_Hiring_Assignment/background_images/BAGGAGE_20180811_175323_83216_B_1.jpg")
bg2 =  Image.open("../BaggageAI_CV_Hiring_Assignment/background_images/BAGGAGE_20180811_175328_83216_A_1.jpg")
bg3 =  Image.open("../BaggageAI_CV_Hiring_Assignment/background_images/S0210209058_20180811232942_L-1_1.jpg")
bg4 =  Image.open("../BaggageAI_CV_Hiring_Assignment/background_images/S0300542812_20180822020845_L-10_1.jpg")
bg5 =  Image.open("../BaggageAI_CV_Hiring_Assignment/background_images/S0320365070_20180821160850_L-12_5.jpg")
```

```python
#cropping threat1
width, height = threat1.size
left = 17
top = height / 4
right = 205
bottom = 3 * height /7
threats1_cropped = threat1.crop((left, top, right, bottom))
  
# Shows the image in image viewer
threats1_cropped.show()
```

```python
#resizing and making it tranparent
img = threats1_cropped
img = img.convert("RGBA")

datas = img.getdata()
  
newData = []
    
for item in datas:
    if item[0] == 255 and item[1] == 255 and item[2] == 255:
        newData.append((255, 255, 255, 0))
   
    else:
        newData.append(item)
  
img.putdata(newData)
   
img.save("./knifeTransparent.png", "PNG")
           
print("Successful")
```
